```python
from typing import Any, TypedDict
from langgraph.graph import StateGraph, START, END
from langgraph.checkpoint.memory import MemorySaver
from IPython.display import Image, display

class SubState(TypedDict):
    sub_output: str


def after(state):
    return {
        "sub_output": "After interruption"
    }


subgraph_builder = StateGraph(SubState)
subgraph_builder.add_node("after", after)
subgraph_builder.add_edge(START, "after")
subgraph_builder.add_edge("after", END)

memory = MemorySaver()

subgraph = subgraph_builder.compile(
    checkpointer=memory, 
    interrupt_before=["after"],
    # debug=True,
    )

display(Image(subgraph.get_graph().draw_mermaid_png()))

thread_id = "3"

class MainState(TypedDict):
    output: str

def wrapper(state: MainState) -> MainState:
    thread = {"configurable": {"thread_id": thread_id}}
    snapshot = subgraph.get_state(config=thread)
    print(f"State: {snapshot.values}")
    print(f"Next node: {snapshot.next}")

    if 'after' in snapshot.next:
        subgraph.invoke(None, thread)
    else:
        subgraph.invoke({"sub_output": "sub"}, thread)
    snapshot = subgraph.get_state(config=thread)
    print(f"New State: {snapshot.values}")
    print(f"Next node: {snapshot.next}")
    return {
        "output": snapshot.values["sub_output"]
    }
    
graph_builder = StateGraph(MainState)
graph_builder.add_node("wrapper", wrapper)
graph_builder.add_edge(START, "wrapper")
graph_builder.add_edge("wrapper", END)

main_graph = graph_builder.compile(
    checkpointer=MemorySaver(),
    # debug=True,
    )

display(Image(main_graph.get_graph().draw_mermaid_png()))

print("RUNNING MAIN GRAPH")

thread = {"configurable": {"thread_id": thread_id}}

output = main_graph.invoke({"output": "first"}, config=thread)

print(f"First output: {output}")

print("---------------")

output = main_graph.invoke({"output": "second"}, config=thread)

print(f"Second output: {output}")

print("---------------")
# print("RUNNING SUBGRAPH")

# thread_id = "5"
# thread = {"configurable": {"thread_id": thread_id}}

# snapshot = subgraph.get_state(config=thread)
# print(f"State: {snapshot.values}")
# print(f"Next node: {snapshot.next}")

# output = subgraph.invoke({"sub_output": "first"}, config=thread)

# print(f"First output: {output}")
# snapshot = subgraph.get_state(config=thread)
# print(f"State: {snapshot.values}")
# print(f"Next node: {snapshot.next}")

# print("---------------")

# output = subgraph.invoke(None, config=thread)

# print(f"Second output: {output}")
# snapshot = subgraph.get_state(config=thread)
# print(f"State: {snapshot.values}")
# print(f"Next node: {snapshot.next}")

# thread_id = "7"

# for event in subgraph.stream({"sub_output": "stream"}, thread, stream_mode="values"):
#     print(event)
```
